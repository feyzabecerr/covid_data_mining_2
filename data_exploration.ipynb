```python
import pandas as pd
import matplotlib.pyplot as plt

#https://www.google.com/covid19/mobility/
df = pd.read_csv('2020_TR_Region_Mobility_Report.csv') #csv dosyasını okuduk
```

```python
df
```

```python
####1.Data üzerinde 5li özet istatiski bilgileri gösteriniz. 
df.describe()
```

```python
####2.Bu datadaki [9-14] sütunlarını birbirlerine karşı scatter plot ile çizerek aralarında herhangi bir
###correlation(positive, negative, none) olup olmadığını belirleyiniz.
fig, ax = plt.subplots(5, figsize=(9, 10))

x1 = df.iloc[:,9]
y1 = df.iloc[:,10]
ax[0].scatter(x = x1, y = y1)
ax[0].set_xlabel("9")
ax[0].set_ylabel("10")

x1 = df.iloc[:,10]
y1 = df.iloc[:,11]
ax[1].scatter(x = x1, y = y1)
ax[1].set_xlabel("10")
ax[1].set_ylabel("11")

x1 = df.iloc[:,11]
y1 = df.iloc[:,12]
ax[2].scatter(x = x1, y = y1)
ax[2].set_xlabel("11")
ax[2].set_ylabel("12")

x1 = df.iloc[:,12]
y1 = df.iloc[:,13]
ax[3].scatter(x = x1, y = y1)
ax[3].set_xlabel("12")
ax[3].set_ylabel("13")

x1 = df.iloc[:,13]
y1 = df.iloc[:,14]
ax[4].scatter(x = x1, y = y1)
ax[4].set_xlabel("13")
ax[4].set_ylabel("14")

plt.show()
```

--> 9-10
-->10-11
Aralarında pozitif korelasyon bulunur.

--> 11-12
--> 12-13
Arasında korelasyon yoktur.  

--> 13-14 
Arasında negatif korelasyon bulunur.

```python
####3.Bu datadaki [9-14] sütunlarını boxplot ile çizirek hangi sütünda outlier olup olmadığını belirleyiniz.

df.iloc[:,9:15].plot(kind='box',title='Boxplot',figsize=(12,8),rot=20,fontsize=8)

# show plot
plt.show()
```

 Tüm sütunlarda outlier bulunur.

```python
####4.Dataframe'i 8.sütun ('date') sütununda aylara göre gruplayarak [9-14] sütunlarını aşağıdakine 
###benzer şekilde plot ediniz (groupby ve DatetimeIndex kullanarak yapabilirsiniz).

df['month'] = pd.DatetimeIndex(df['date']).month  
df['date'] = df['month'] # tablodaki date değerlerini month değerleriyle değiştirdik
# plot data
fig, ax = plt.subplots(figsize=(12,8))
df.iloc[:,8:15].groupby(['date']).mean().plot(ax=ax)
```

```python
####5. https://drive.google.com/file/d/1Eg8Lffm49bc-bGFkv_4ddrQw8U8WE6P4/view?usp=sharing adresinde 2021 TR datası vardır.
### Bu datayıda 4.deki gibi gruplayarak, her iki datayı karşılaştırmak için aynı indeksli sütunlarını aylara göre aynı plotta gösteriniz.

df2 = pd.read_csv('2021_TR_Region_Mobility_Report.csv') # csv dosyasını okuduk
df2
```

```python
df2['month'] = pd.DatetimeIndex(df2['date']).month  
df2['date'] = df2['month'] # tablodaki date değerlerini month değerleriyle değiştirdik
# plot data
fig, ax = plt.subplots(figsize=(12,8))
df2.iloc[:,8:15].groupby(['date']).mean().plot(ax=ax)
```

```python
# Plotları yan yana çizdirme

fig, (ax, ax2)  = plt.subplots(ncols=2,figsize=(20,7))

df.iloc[:,8:15].groupby(['date']).mean().plot(ax=ax)
df2.iloc[:,8:15].groupby(['date']).mean().plot(ax=ax2, ls = "--")
plt.show()
```

```python
# Plotları aynı grafikte gösterme

fig, ax = plt.subplots(figsize=(15,7))
ax2 = ax.twinx()

df.iloc[:,8:15].groupby(['date']).mean().plot(ax=ax)
df2.iloc[:,8:15].groupby(['date']).mean().plot(ax=ax2, ls = "--")
plt.show()
```
